```python
import pandas as pd
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns

matplotlib.style.use('ggplot')

from bokeh.plotting import figure, show
from bokeh.io import output_notebook

output_notebook()
% matplotlib inline

```

```python
# Vizualização de Dados é muito importante. Pode nos revelar coisas sobre os dados de maneira mais clara e direta.


df1 = pd.DataFrame({'x':[10.0, 8.0, 13.0, 9.0, 11.0, 14.0, 6.0, 4.0, 12.0, 7.0, 5.0],
                   'y':[8.04, 6.95, 7.58, 8.81, 8.33, 9.96, 7.24, 4.26, 10.84, 4.82, 5.68]})
df2 = pd.DataFrame({'x':[10.0, 8.0, 13.0, 9.0, 11.0, 14.0, 6.0, 4.0, 12.0, 7.0, 5.0],
                   'y':[9.14, 8.14, 8.74, 8.77, 9.26, 8.10, 6.13, 3.10, 9.13, 7.26, 4.74]})
df3 = pd.DataFrame({'x':[10.0, 8.0, 13.0, 9.0, 11.0, 14.0, 6.0, 4.0, 12.0, 7.0, 5.0],
                   'y':[7.46, 6.77, 12.74, 7.11, 7.81, 8.84, 6.08, 5.39, 8.15, 6.42, 5.73]})
df4 = pd.DataFrame({'x':[8.0, 8.0, 8.0, 8.0, 8.0, 8.0, 8.0, 19.0, 8.0, 8.0, 8.0],
                   'y':[6.58, 5.76, 7.71, 8.84, 8.47, 7.04, 5.25, 12.50, 5.56, 7.91, 6.89]})

df = pd.concat([df1,df2,df3,df4], axis=1, 
               keys=list(map(lambda x: 'df{}'.format(x), range(1,5))))

df
```

```python
# Apesar dos dados serem diferentes, eles possuem as mesmas medidas estatísticas

f = lambda x: pd.Series({'media':np.round(x.mean(),1),'desv_pad':np.round(x.std(),1)})
pd.concat([df1.apply(f),df2.apply(f),df3.apply(f),df4.apply(f)],axis=1,
          keys=list(map(lambda x: 'df{}'.format(x), range(1,5))))

```

```python
# Mas com vizualização vemos como são diferentes

fig, axes = plt.subplots(ncols=2,nrows=2,figsize=(12,5))
axes[0][0].set_xlim([0,20]);axes[0][0].set_ylim([0,14]);axes[0][1].set_xlim([0,20]),axes[0][1].set_ylim([0,14]);axes[1][0].set_xlim([0,20]);axes[1][0].set_ylim([0,14]);axes[1][1].set_xlim([0,20]);axes[1][1].set_ylim([0,14])
sns.regplot(x='x',y='y',data=df1,ci=None,ax=axes[0][0])
sns.regplot(x='x',y='y',data=df2,ci=None,ax=axes[0][1])
sns.regplot(x='x',y='y',data=df3,ci=None,ax=axes[1][0])
sns.regplot(x='x',y='y',data=df4,ci=None,ax=axes[1][1]);
```

```python
# Barras

BRICS = ["Brazil", "Russian Federation","India", 'China', 'South Africa', 'Euro area', 'United States']
gdp.loc[BRICS,'2010'].plot.bar()

gdp.loc[BRICS,['1990','1995','2000','2005','2010','2014']].T.plot.bar();

gdp.loc[BRICS,['1990','1995','2000','2005','2010','2014']].plot.bar(subplots=True,
                                                                    layout=(2,3),
                                                                    sharey=True,
                                                                   figsize=(12,7));
```

```python
# Histograma

fig, ax = plt.subplots()
ax.set_xlim([0,100])
expecVida['2010'].plot.hist(bins=30);

expecVida[['1960','1980','2000','2014']].plot.hist(stacked=False,
                                                 bins=20, alpha=0.4);

```

```python
# Densidade
expecVida[['1960','1980','2000','2014']].plot.density(alpha=.8);
```

```python
# Box Plot
expecVida.loc[BRICS].T.plot.box(rot=90,sym='*',showmeans=True,meanline=True);

# O Box Plot pode ser confuso, mas não é tanto, para entende-lo, basta sacar oque ele representa
# que são aqueles 5 dados importante do describe. A linha do meio da caixa é a mediana, embaixo dela temos o primeiro quartil
# e emcima o terceiro quartil. A linha pontilhada é a média, e os bigodes são determinados pela seguinte formula:
# Bigode de cima : Menor valor entre o máximo e Q3 + 1.5 x IQR(Interquartil amplitude)
# Bigode de baixo: Maior valor entre o mínimo e Q1 - 1.5 x IQR
# Basicamente entre os bigodes temos 99% dos dados, e oque não está lá teoricamente pode ser descartado. São as estrelinhas
```

```python
# Gráficos de Dispersão

df = pd.melt(gdp.reset_index(),id_vars='Country Name'); df.set_index(['Country Name'], inplace=True); 
df.variable = df.variable.astype('int64')
df.loc['Brazil'].plot.scatter(x='variable',y='value');

fig, ax = plt.subplots(ncols=2,figsize=(12,5))
df.loc['Brazil'].plot.scatter(x='variable',y='value',ax=ax[0])
df = pd.melt(expecVida.reset_index(),id_vars='Country Name')
df.set_index(['Country Name'], inplace=True)
df.variable = df.variable.astype(int)
df.loc['Brazil'].plot.scatter(x='variable',y='value',ax=ax[1]);

```
